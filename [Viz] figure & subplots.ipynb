### subplot( ) function

#### 1) figure for ax1 and ax2

```python
fig = plt.figure()

ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)

plt.show()
```

### More example

```python
import matplotlib.pyplot as plt

fig = plt.figure()
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)

x = range(0,4)      
y = [v*v for v in x]

ax1.plot(x,y)
ax2.bar(x,y)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0.0, 2 * np.pi, 0.1)
sin_y = np.sin(x)
cos_y = np.cos(x)

fig = plt.figure(figsize=(12,3))   # figure size

ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

ax1.plot(x, sin_y, 'b--')
ax2.plot(x, cos_y, 'r--')

plt.show()
```

## subplot & subplots

#### subplot
 - subplot : 일일이 다 설정해줘야 함

```python
import matplotlib.pyplot as plt

fig = plt.figure()  

plt.subplot(2,1,1)
plt.subplot(2,1,2)

plt.show()
```

#### subplots
 - 한 번에 설정해줄 수 있음

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(nrows=2, ncols=1)

plt.show()
```

#### fig, ax 의미

```python
fig, ax = plt.subplots()
```

 - subplots()에서는 2개의 값을 받을 수 있음.
 - 2개 : figure & axes
 - figure(fig) : 전체 subplot들을 담는 하나의 영역. 전체 사이즈.
 - axes(ax) : 전체 중 낱낱개의 subplot들

```python
print(fig)
```

```python
print(ax)
```

반면, subplot()의 경우는 하나의 값(axes)만 받을 수 있음.

```python
ax = plt.subplot(2,1,1)
```

#### subplots
 - matplotlib.pyplot.subplots(nrows=1, ncols=1, sharex=False, sharey=False, squeeze=True, subplot_kw=None, gridspec_kw=None, **fig_kw)
   - nrows : 몇 행을 만들 것인지
   - ncols : 몇 열을 만들 것인지
   - sharex : Xtick의 값을 각각 공유할지 안 할지
   - sharey : Ytick의 값을 각각 공유할지 안 할지

## 해석해보기

```python
size_graph = 5

for idx_x in range(size_graph):
    
    fig, axes = plt.subplots(1, 2, figsize=(25, 3))
    
    axes[0].plot(np.random.rand(100))
    axes[0].set_ylim(0, 1.2)
    axes[0].set_ylabel('Probability', fontsize=20)
    # axes[0].grid()
    
    axes[1].plot(np.random.rand(100))
    axes[1].set_ylim(0, 2.4)
    axes[1].set_ylabel('Probability', fontsize=20)
    axes[1].grid()

plt.show()
```

#### step by step

```python
fig, axes = plt.subplots(1, 2, figsize=(25, 3))
fig, axes = plt.subplots(1, 2, figsize=(25, 3))
fig, axes = plt.subplots(1, 2, figsize=(25, 3))
fig, axes = plt.subplots(1, 2, figsize=(25, 3))
fig, axes = plt.subplots(1, 2, figsize=(25, 3))
```

## plt.axis()

```python
plt.axis('on')   # 축과 이름을 보여줌 / default
```

```python
plt.axis('off')   # 축과 이름을 제거
```

```python
plt.axis('equal')  # x축과 y축의 비율을 실제 똑같은 비율로 나오도록 함
```

```python
plt.axis('scaled')   # x축과 y축의 비율을 실제 똑같은 비율로 나오게 함. & 여백을 축소
```
