Conditional Expression

```python
x = 5 
y = x if x==10 else 0
y
```

```python
if True:
    print('true_1')
else:
    print('false_1')

if False:
    print('true_2')
else:
    print('false_2')
    
if None:
    print('true_3')
else:
    print('false_3')
```

```python
if 0 : print('true_1') 
else : print('false_1')
    
if 1 : print('true_2') 
else : print('false_2')

if 13.5 : print('true_3') 
else : print('false_3')

if 0x1F : print('true_4') 
else : print('false_4')

if 0b1000 : print('true_5') 
else : print('false_5')
```

```python
if 'hi' : print('true_1') 
else : print('false_1')
    
if '' : print('true_2') 
else : print('false_2')
```

#### Summary
 - return True : numbers(except 0), strings,...
 - return False : 0, None, '', [], (), {}, SET() ... 

```python
a,b,c,d = map(int, input().split())
if 0<=a<=100 and 0<=b<=100 and 0<=c<=100 and 0<=d<=100 :    
    if (a+b+c+d)/4 >= 80:
        print('pass')
    else:
        print('non-pass')
else :
    print('wrong input')
```

```python
if 7<=age<=12 :
    balance = balance - 650
elif 13<=age<=18 :
    balance = balance - 1050
elif age>=19:
    balance = balance - 1250

print(balance)
```

#### for phrase
 - range / increase values / decrease values

```python
for i in range(0,10,2):
    print('hi,increasing',i)
```

```python
for i in range(10,0,-2):
    print('hi, decreasing',i)
```

reversed 이용

```python
for i in reversed(range(5)):
    print('hi',i)
```

```python
for i in range(10):
    print(i, end=' ')
    i = 10   # no effect
```

use sequence object(list, tuple, ...) into for phrase

```python
a = [1,2,3,4,5]
for i in a:
    print(i,end=' ')
```

```python
a = ('aaa','bb','c')
for i in a:
    print(i,end=' ')
```

```python
a = 'python'
for i in a:
    print(i,end=' ')
```

```python
for letter in reversed('python'):
    print(letter, end=' ')
```

#### while phrase

```python
i = 0            # i=1
while i<10:      # i<=10
    print('hi',i,end='/')
    i += 1
```

**random.choice** : select element randomly from sequence object

```python
dice = [1,2,3,4,5,6]
print(random.choice(dice))
print(random.choice(dice))
```

#### break

```python
i = 0
while True:
    print(i, end=' ')
    i += 1
    if i == 5:
        break
```

```python
for i in range(10000):
    print(i, end=' ')
    if i == 5:
        break 
```

#### continue

```python
for i in range(8):
    if i%2 == 0:
        continue   # skip 
    print(i, end=' ')
```

```python
i = 0
while i < 10:
    i+=1
    if i%2 == 0:
        continue
    print(i, end=' ')
```
