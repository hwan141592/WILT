#### nested loop

```python
for i in range(5):
    for j in range(5):
        print('j:',j,sep='',end=' ')
    print('i:',i,'\\n',sep='')
```

```python
for i in range(3):
    for j in range(5):
        print("*",end=' ')
    print()
```

```python
for i in range(5):
    for j in range(5):
        if j<=i :
            print("*",end=' ')
    print()
```

```python
for i in range(5):
    for j in range(5):
        if j==i:
            print('*', end='')
        else:
            print(' ', end='')
    print()
```

```python
for i in range(5):
    for j in range(5):
        if j<i:
            print(' ', end='')
        else:
            print('*', end='')            
    print()
```

```python
cnt = 5
for i in range(cnt):
    for j in reversed(range(cnt)):
        if j>i :
            print(' ', end='')
        else:
            print('*',end='')
    for j in range(cnt):
        if  j<i :
            print('*',end='')
    print()
```

### FizzBuzz Question
 - First, prints each number from 1 to 100 on a new line. 
 - For each multiple of 3, print "Fizz" instead of the number. 
 - For each multiple of 5, print "Buzz" instead of the number. 
 - For numbers which are multiples of both 3 and 5, print "FizzBuzz" instead of the number.

```python
for i in range(1,101):
    if i%3 ==0 and i%5 == 0:
        print('FizzBuzz', end=' ')
    elif i%3 == 0:
        print('Fizz', end=' ')
    elif i%5 == 0:
        print('Buzz', end=' ')
    else:
        print(i, end=' ')
```

#### other solution

```python
for i in range(1,101):
    if i%15 == 0 :  
        print('FizzBuzz', end=' ')
    elif i%3 == 0:
        print('Fizz', end=' ')
    elif i%5 == 0:
        print('Buzz', end=' ')
    else:
        print(i, end=' ')
```

#### code golf

```python
for i in range(1,50):
    print('Fizz' * (i%3==0) + 'Buzz' * (i%5==0) or i)
```
