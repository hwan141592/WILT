```python
import numpy as np
import matplotlib.pyplot as plt
```

# # Scatter Plot
 - https://matplotlib.org/3.5.0/api/_as_gen/matplotlib.pyplot.scatter.html  
 - matplotlib.pyplot.plt.scatter(x, y, ...)
   - x: float or array-like, shape (n, )
   - y: float or array-like, shape (n, )


### ### Plotting

```python
# list
x = [1, 5, 4, 2, 3]
y = [100, 200, 150, 115, 210]
plt.figure(figsize=(12,4))
plt.scatter(x, y)
plt.show()
```

```python
# tuple
x = (1, 5, 4, 2, 3)
y = (100, 200, 150, 115, 210)
plt.figure(figsize=(12,2))
plt.scatter(x, y)
plt.show()
```

```python
# dictionary
dic = {1:100, 5:200, 4:150, 2:115, 3:210}
plt.figure(figsize=(12,4))
plt.scatter(dic.keys(), dic.values())
plt.show()
```

```python
# nd array
x = np.array([1, 5, 4, 2, 3])
y = np.array([100, 200, 150, 115, 210])
plt.figure(figsize=(12,2))
plt.scatter(x, y)
plt.show()
```

### ### Visualization

#### 1) x: numerical, y: numerical

```python
x = np.array([1, 5, 4, 2, 3])
y = np.array([100, 200, 150, 115, 210])

arr_size = [50, 250, 100, 200, 160]
arr_color = ['orange', 'yellow', 'red', 'blue', 'green']

plt.figure(figsize=(12,2))
plt.scatter(x, y, s=arr_size, c=arr_color)

plt.title('Title of the visualization', fontsize=20)
plt.xlabel('x-axis name', fontsize=14)
plt.ylabel('y-axis name', fontsize=14)

plt.xlim(0, 6)
plt.ylim(0, 300)
plt.xscale('linear')   # linear, log, symlog, logit 

plt.xticks(rotation=45)
# plt.xticks([])   # Disable xticks.

# plt.xticks(np.arange(0, 1, step=0.2))             # Set label locations.
# plt.xticks(np.arange(3), ['Tom', 'Dick', 'Sue'])  # Set text labels.
# plt.xticks([0, 1, 2], ['January', 'February', 'March'], rotation=20)  

plt.show()
```

#### 2) x: categorical, y: numerical

```python
x = np.array(['x1', 'x2', 'x5', 'x3', 'x4'])
y = np.array([100, 200, 150, 115, 210])

arr_size = [50, 250, 100, 200, 160]
arr_color = ['orange', 'yellow', 'red', 'blue', 'green']

plt.figure(figsize=(12,2))
plt.scatter(x, y, s=arr_size, c=arr_color)

plt.title('Title of the visualization', fontsize=20)
plt.xlabel('x-axis name', fontsize=14)
plt.ylabel('y-axis name', fontsize=14)

plt.xticks(rotation=45)
# plt.xticks([])   # Disable xticks.

plt.ylim(50, 250)
plt.show()
```
