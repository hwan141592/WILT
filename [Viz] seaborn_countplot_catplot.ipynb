```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

## seaborn.countplot
 - https://seaborn.pydata.org/generated/seaborn.countplot.html

```python
sns.set_style(style='darkgrid')
titanic = sns.load_dataset('titanic')
titanic.head(3)
```

```python
ax = sns.countplot(x='class', data=titanic)
```

```python
ax = sns.countplot(y='class', data=titanic)
```

```python
ax = sns.countplot(x='class', hue='who', data=titanic)
```

```python
ax = sns.countplot(y='class', hue='who', data=titanic)
```

```python
ax = sns.countplot(x='class', data=titanic, palette='Set3')
```

```python
ax = sns.countplot(x = 'class', 
                   data = titanic,
                   facecolor = (0, 0, 0, 0),
                   linewidth = 5,
                   edgecolor = sns.color_palette('dark', 3)
                  )
```

## seaborn.catplot

```python
g = sns.catplot(x = 'class', data = titanic, kind = 'count',
                height = 3, aspect = 1.8)
```

```python
g = sns.catplot(x='class', hue='who', col='survived',
                data=titanic, kind='count',
                height=3, aspect=.9)
```

```python
import pandas as pd
import seaborn as sns

li_x = ['a', 'b', 'c', 'd', 'e']
li_y = [80, 90, 100, 95, 75]
df = pd.DataFrame({'col_x':li_x, 'col_y':li_y})
display(df, sns.catplot(x='col_x', y='col_y', data=df, kind='bar', height=4, aspect=2))
```

```python
import pandas as pd
import seaborn as sns

li_x = ['a', 'a', 'a', 'b', 'b', 'b']
li_y = [80, 90, 100, 85, 70, 55]
df = pd.DataFrame({'col_x':li_x, 'col_y':li_y})
display(df, df.groupby('col_x').mean())
```

```python
sns.catplot(x='col_x', y='col_y', data=df, kind='bar', height=4, aspect=2)   # default : average
```
