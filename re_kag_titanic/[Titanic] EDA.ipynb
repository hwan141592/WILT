### [feature.06] Fare
 - fare for aboarding
 - continuous feature

```python
f, ax = plt.subplots(1,1,figsize=(8,8))

g = sns.distplot(df_train['Fare'], color='b', label='Skewness: {:.2f}'.format(df_train['Fare'].skew()), ax=ax)
g = g.legend(loc='best')
```

 - distribution of fare is very asymmetric  
 - maybe a problem when training, since model can be affected so sensitively by outlier  
 - to reduce effect of outlier, let's take log to fare (using pandas apply,map, etc)
 - before taking log, fill nan value as mean value of fare

```python
df_test.loc[df_test.Fare.isnull(), 'Fare'] = df_test['Fare'].mean()
df_train['Fare'] = df_train['Fare'].map(lambda i: np.log(i) if i>0 else 0)
df_test['Fare'] = df_test['Fare'].map(lambda i: np.log(i) if i>0 else 0)
```

```python
fig, ax = plt.subplots(1,1,figsize=(8,8))
g = sns.distplot(df_train['Fare'], color='b', label='Skewness: {:.2f}'.format(df_train['Fare'].skew()), ax=ax)
g = g.legend(loc='best')
```

 - After taking log to fare column, asymmetric problem seems to be solved  
   (one part of feature engineering)
